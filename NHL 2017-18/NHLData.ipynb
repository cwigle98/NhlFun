```python
import pandas as pd
```

```python
#Create functions to use with data
##Loads in the data table for the functions
def data():
    stats = pd.read_csv('DataTables/playerData.csv')
    stats.iloc[0:,1] = list(map(lambda x: x.split('\\')[0],stats.iloc[0:,1]))
    stats = stats[stats['Team'] != 'TOT']
    stats = stats.drop(['Rk'], axis=1)
    return stats
##Prints the roster of each team
    ##key: The name of the team you are searching for
    ##return the roster of a team in a pandas data table
def Roster(key):
    stats = data()
    teamDict = createCityKey()
    teams = list(teamDict.keys())
    tem = 0
    if(key in 'Vegas Golden Knights'):
        print('Team data not yet available')
    else:
        for i in range(0,len(teams)):
            if(key.lower() in teams[i].lower()):
                tem = teams[i]
        if(tem == 0):
            print('Could not find team')
        else:
            return stats[stats['Team'] == teamDict[tem]]
##Creates a dictionary of team names to their abbreviations
    ##return a dictionary of team names to their abbreviations
def createCityKey():
    teams = pd.read_excel('DataTables/teamList.xlsx')
    teamDict = {}
    for i in range(0,len(teams.iloc[0:,0])):
        teamDict[teams.iloc[i,0]] = teams.iloc[i,1]
    return teamDict
##Sorts the data based on the key
    ##key:The stat to sort by
    ##ascending
def sort(key,ascending = None):
    stats = data()
    if(ascending == None):
        ascending == False
    return stats.sort_values(key,ascending=ascending)
def stats():
    stats = data()
    return pd.DataFrame(stats.columns)
def sortTeam(team,key,ascending = None):
    stats = Roster(team)
    if(ascending == None):
        ascending == False
    return stats.sort_values(key,ascending=ascending)
```
